## 重力位能的畫圖——航海家一號

目前人類發射的最遠的飛行器是美國航空總署在 1977 年發射的航海家一號（Voyage 1）。它現在已經距離太陽 $2.19\times10^{10} \ \text{m}$ 了，是在離太陽系十分遙遠的星際空間！

航海家一號在 1979 年 3 月 5 日飛掠了木星，當時的距離是 $349,000 \ \text{km}$。

記得一個物體在地球重力下的位能公式為

$$V(r) = -\frac{GMm}{r},$$

其中的 $G=6.674\times10^{-11}\ \text{m}^3\text{kg}^{-1}\text{s}^{-2}$ 為萬有引力常數，$M=5.97\times10^{24}\ \text{kg}$ 為地球質量，而 $m$ 為物體質量。

已知航海家一號的質量為 $825.5 \ \text{kg}$。試著用 `numpy` 畫出航海家一號從地表（$6317\times10^3 \ \text{m}$）出發後到飛掠木星所經歷的地球位能吧！



提示：將 x 軸設定成 $6317\times10^3 \ \text{m} \leq x \leq 349\times10^6 \ \text{m}$。

```python
x = np.linspace(6317e3, 349e6, 10000)
G = 6.674e-11
M = 5.97e24
m = 825.5
y = - (G * M * m) / x
plt.plot(x, y)
```
